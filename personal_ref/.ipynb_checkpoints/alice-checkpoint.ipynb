## Sampling out 10 random, with quality good, and T-stat > 3 (statistical significance)


```python
signal_df['T-Stat'] = pd.to_numeric(signal_df['T-Stat'], errors='coerce')
good_signals_df = signal_df[signal_df['Signal Rep Quality'] == '1_good']

strong_signals_df = good_signals_df[good_signals_df['T-Stat'] > 3]

sampled_strong_signals = strong_signals_df#.sample(n=40, random_state=42)
sampled_strong_acronyms = sampled_strong_signals['Acronym'].tolist()

sampled_strong_signals
filtered_signals_reset = sampled_strong_signals.reset_index(drop=True)
filtered_signals_reset[['Acronym', 'Signal Rep Quality', 'Return', 'T-Stat']]

```

```python
signals = filtered_signals_reset['Acronym'].tolist() 
```

```python
port_vw = openap.dl_port('deciles_vw', 'pandas', signals)
ls_df = port_vw[port_vw['port'] == 'LS']
```

```python
ls_df
```

```python
ls_matrix = ls_df.pivot(index='date', columns='signalname', values='ret')
ls_matrix
```

```python
ls_zscore = (ls_matrix - ls_matrix.mean()) / ls_matrix.std()
ls_zscore.dropna()
```

```python
weights = filtered_signals_reset.set_index('Acronym')['T-Stat']
weights_aligned = weights.reindex(ls_zscore.columns).fillna(0)
weights_normalized = weights_aligned / weights_aligned.sum()
composite_signal = ls_zscore.dot(weights_normalized)
#composite_signal = composite_signal.dropna()
```

```python
actual_ls_return = ls_matrix.mean(axis=1)
aligned = pd.concat([composite_signal, actual_ls_return.shift(-1)], axis=1).dropna()
aligned.columns = ['predicted', 'actual']

print(aligned.corr())
```

```python
algos = [linear_r, rf, xgb_model, svr, MLP]
X = ls_zscore.shift(1).interpolate().dropna()

y = ls_matrix.mean(axis=1).loc[X.index] 
print(X.isna().sum())
```

```python
def summary_stats(series):
    mean = series.mean()
    std = series.std()
    sharpe = mean / std
    t_stat = mean / (std / np.sqrt(len(series)))
    print(f"  Mean Return: {mean:.4f}")
    print(f"  Volatility: {std:.4f}")
    print(f"  Sharpe Ratio: {sharpe:.4f}")
    print(f"  T-Statistic: {t_stat:.4f}")
    print(f"  Start Date: {series.index.min()}")
    print(f"  End Date: {series.index.max()}")
    print("-" * 50)
    return " "

summary_stats(composite_signal)
```

```python
composite_cumulative_return = composite_signal.cumsum()
actual_cumulative_return = y.cumsum()
composite_cumulative_return = composite_cumulative_return.interpolate()
actual_cumulative_return = actual_cumulative_return.interpolate()


plt.figure(figsize=(10, 6))
plt.plot(composite_cumulative_return.index, composite_cumulative_return, label="Composite Signal", color='blue', linestyle='-', linewidth=2)
plt.plot(actual_cumulative_return.index, actual_cumulative_return, label="Actual", color='green', linestyle=':', linewidth=2)
plt.title("Cumulative Return: Composite Signal vs Actual Values")
plt.xlabel("Date")  # or "Index" if you're using indices
plt.ylabel("Cumulative Return")
plt.legend(loc='best')

plt.grid(True)
plt.show()

composite_cumulative_return, actual_cumulative_return = composite_cumulative_return.align(actual_cumulative_return, join='inner')
correlation = composite_cumulative_return.corr(actual_cumulative_return)
mse = mean_squared_error(actual_cumulative_return, composite_cumulative_return)

print(f"Correlation: {correlation}")
print(f"Mean Squared Error: {mse}")
```

```python
plt.figure(figsize=(10, 6))

colors = {
    "LinearRegression": 'blue',
    "RandomForestRegressor": 'orange',
    "XGBRegressor": 'red',
    "SVR": 'black',
    "MLPRegressor": 'yellow'
}
for algo in algos:
    algo.fit(X, y)
    y_test_pred = pd.Series(algo.predict(X), index=X.index)
    print(type(algo).__name__)
    print(f"  Mean Squared Error: {mean_squared_error(y, y_test_pred)}")
    print(f"  Correlation: {y_test_pred.corr(y)}")
    print(summary_stats(y_test_pred))
    plt.plot(y_test_pred.index, y_test_pred.cumsum(), label=f"Predicted - {type(algo).__name__}", color=colors[type(algo).__name__])

plt.plot(y.index, y.cumsum(), label="Actual", color='green', linewidth=2, linestyle=':')
plt.title("Cumulative Actual vs Predicted Values for All Models")
plt.xlabel("Index")
plt.ylabel("Cumulative Value")
plt.legend()  
plt.grid(True) 
plt.show()
```

```python
crsp_df['date'] = pd.to_datetime(crsp_df['date'])   
crsp_df['yyyymm'] = crsp_df['date'].dt.year * 100 + crsp_df['date'].dt.month  

crsp_df['permno'] = crsp_df['permno'].astype('int64')
crsp_df['yyyymm'] = crsp_df['yyyymm'].astype('int64')

allsignal_20['permno'] = allsignal_20['permno'].astype('int64')
allsignal_20['yyyymm'] = allsignal_20['yyyymm'].astype('int64')

merged_df = pd.merge(allsignal_20, crsp_df, on=["permno", "yyyymm"], how="inner")
merged_df = merged_df.dropna(subset=["ret"])
merged_df
```

```python
import pandas as pd
from lifelines import CoxPHFitter

# Step 1: Define Death (return <= -50% over next 6 months)
merged_df = merged_df.sort_values(by=['permno', 'yyyymm'])

# Rolling 6-month return
merged_df['future_ret_6m'] = merged_df.groupby('permno')['ret'].rolling(window=6, min_periods=1).sum().shift(-6).reset_index(level=0, drop=True)

# Event: 1 if future 6M return <= -50%
merged_df['event'] = (merged_df['future_ret_6m'] <= -0.5).astype(int)

# Step 2: Duration
# For now, simple duration of 1 (because each row is one monthly observation)
merged_df['duration'] = 1

survival_df = merged_df[["duration", "event"] + top_20_signals].dropna()

# Step 4: Train Cox Proportional Hazards model
cph = CoxPHFitter()
cph.fit(survival_df, duration_col='duration', event_col='event')


# Step 5: Summary Output
cph.print_summary()  # Hazard Ratios, p-values, confidence intervals
```

```python
# 1. Get summary with index reset
summary_df = cph.summary.reset_index()

# 2. Rename the index column to "Acronym" (whatever it's called — find out)
summary_df.columns = ['Acronym' if col == summary_df.columns[0] else col for col in summary_df.columns]

# 3. Map the detailed definition from another dataframe
# Make sure signal_df has 'Acronym' and 'Detailed Definition' columns
summary_df['Detailed Definition'] = summary_df['Acronym'].map(
    signal_df.set_index('Acronym')['Detailed Definition']
)

# 4. (Optional) Move 'Detailed Definition' next to Acronym for display
cols = ['Acronym', 'Detailed Definition'] + [col for col in summary_df.columns if col not in ['Acronym', 'Detailed Definition']]
summary_df = summary_df[cols]

summary_df
```

```python
# --- Signal Engineering Section ---
# Create engineered features based on combinations of top 20 signals

# Interaction terms
allsignal_20['XFIN_NetEquityFinance'] = allsignal_20['XFIN'] * allsignal_20['NetEquityFinance']
allsignal_20['TrendFactor_TotalAccruals'] = allsignal_20['TrendFactor'] * allsignal_20['TotalAccruals']

# Ratios
allsignal_20['NetEquityFinance_to_TotalAccruals'] = allsignal_20['NetEquityFinance'] / (allsignal_20['TotalAccruals'].replace(0, np.nan))
allsignal_20['RDS_to_InvGrowth'] = allsignal_20['RDS'] / (allsignal_20['InvGrowth'].replace(0, np.nan))

# Moving averages (within each stock)
allsignal_20 = allsignal_20.sort_values(['permno', 'yyyymm'])
for feature in top_20_signals:
    allsignal_20[f'{feature}_MA3'] = allsignal_20.groupby('permno')[feature].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
    allsignal_20[f'{feature}_MA6'] = allsignal_20.groupby('permno')[feature].transform(lambda x: x.rolling(window=6, min_periods=1).mean())

# Example: Feature cross-products or higher-order terms if you want
allsignal_20['XFIN_squared'] = allsignal_20['XFIN'] ** 2

# After doing the feature engineering step
merged_df = pd.merge(allsignal_20, crsp_df, on=["permno", "yyyymm"], how="inner")

# Drop rows where "ret" is missing (as you did)
merged_df = merged_df.dropna(subset=["ret"])

```

```python
engineered_features = [col for col in allsignal_20.columns if any(keyword in col for keyword in ['_', 'MA3', 'MA6'])]

X_eng = merged_df[engineered_features].dropna()
y_eng = merged_df.loc[X_eng.index, 'ret']

rf.fit(X_eng, y_eng)
importance_eng = pd.Series(rf.feature_importances_, index=X_eng.columns).sort_values(ascending=False)

print("Top Engineered Features by Importance:")
print(importance_eng.head(10))
```

```python
# Use your clean top 20 signals directly
original_features = top_20_signals

# Build X and y
X_orig = merged_df[original_features].dropna()
y_orig = merged_df.loc[X_orig.index, 'ret']

# Fit random forest
rf.fit(X_orig, y_orig)

# Get feature importance
feature_importance_orig = rf.feature_importances_

# Create DataFrame
importance_df_orig = pd.DataFrame({
    'Feature': X_orig.columns,
    'Importance': feature_importance_orig
})

# Sort and print top 10
importance_df_orig = importance_df_orig.sort_values(by='Importance', ascending=False).head(10)

print("Top 10 Original Features by Importance:")
print(importance_df_orig)
```

In this project, I began by downloading a broad universe of financial signals from OpenAssetPricing. I filtered the dataset to retain only signals that had strong historical predictive power, specifically those with a T-statistic greater than 3 and high replication quality. After selecting the strongest signals, I retrieved portfolio returns data and constructed a composite signal. I standardized the signals using z-scores, weighted them according to their T-statistics to prioritize higher-quality predictors, and combined them into a single composite score. To evaluate its effectiveness, I measured the correlation between the composite signal and future returns, computed key summary statistics like the Sharpe ratio and t-statistic, and plotted cumulative returns over time. I found that the composite signal showed a positive relationship with future returns, although some periods demonstrated stronger predictability than others.

Next, I implemented several machine learning models to predict future returns based on the signals, including random forests, gradient boosting (XGBoost), support vector regression (SVR), and multilayer perceptrons (neural networks). For each model, I trained on the z-scored signals and evaluated performance using metrics such as mean squared error (MSE) and out-of-sample correlation. I also plotted cumulative predicted returns for each model against the realized average returns to assess how closely each model tracked actual market outcomes. Random forests and gradient boosting models generally performed best in terms of stability and predictive accuracy. To better understand model behavior, I extracted feature importances from the tree-based models, revealing which signals the models considered most influential in driving future returns.

After establishing a baseline with the original signals, I moved into signal engineering to create new, enhanced features. I engineered new signals by creating interaction terms between existing factors, constructing ratios, and applying moving averages (3-month and 6-month smoothing) to reduce noise and better capture underlying trends. Once I expanded the feature set, I retrained the random forest exclusively on the engineered features. I identified the top 10 most important engineered predictors, many of which were moving averages of original signals such as retConglomerate_MA3 and CustomerMomentum_MA3. Separately, I also retrained a random forest on only the original signals to extract the top 10 original features by importance.

Comparing the two groups revealed several important insights. First, the original signal retConglomerate was by far the strongest predictor in its raw form, accounting for nearly 27% of the model's decision-making weight. However, many momentum- and volatility-related signals such as CustomerMomentum, MomSeason16YrPlus, IndMom, and betaVIX improved significantly after being smoothed through moving averages. The engineered features not only preserved the predictive power of the strongest signals but also created a healthier distribution of importance across a wider set of predictors. This led to a more balanced model that did not overly rely on a single factor, which could improve stability and robustness in out-of-sample tests.

Overall, my analysis suggests that signal engineering materially enhances the predictive power of certain types of signals, especially momentum, trend-following, and volatility exposure variables. Moving averages were particularly effective in reducing short-term noise while preserving directional information. While a few raw signals like retConglomerate were already extremely strong without any modifications, most signals benefited from additional feature engineering. This highlights the importance of not only selecting strong fundamental predictors but also thoughtfully transforming them to maximize their signal-to-noise ratio. In a practical investment strategy, a hybrid model blending both top raw signals and top engineered signals would likely outperform models based on either type alone.
