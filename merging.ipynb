import pandas as pd

# Step 1: Fix crsp's date
crsp['date'] = pd.to_datetime(crsp['date'])   # Properly parse to datetime
crsp['yyyymm'] = crsp['date'].dt.year * 100 + crsp['date'].dt.month  # Create yyyymm column

# Step 2: Make sure permno and yyyymm are int64
crsp['permno'] = crsp['permno'].astype('int64')
crsp['yyyymm'] = crsp['yyyymm'].astype('int64')

df['permno'] = df['permno'].astype('int64')
df['yyyymm'] = df['yyyymm'].astype('int64')

# Step 3: Now merge correctly
merged_df = pd.merge(crsp, df, on=["permno", "yyyymm"], how="inner")

# Step 4: Drop rows where 'ret' is missing
merged_df = merged_df.dropna(subset=["ret"])
